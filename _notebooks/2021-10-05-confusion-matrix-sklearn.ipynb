```python
#collapse-hide
import numpy as np
from sklearn.metrics import confusion_matrix
```

## The confusion matrix
We need some actual and some predicted values.

```python
y_true = np.random.choice(['healthy', 'sick'], 10)
y_true
```

```python
y_pred = np.random.choice(['healthy', 'sick'], 10)
y_pred
```

With these two arrays, we are able to build the confusion matrix with entries

$$
\begin{bmatrix}
TN & FP\\
FN & TP
\end{bmatrix}
$$

TN: True negatives
FN: False negatives
TP: True positives
FP: False positives


```python
confusion_matrix(y_true, y_pred, labels=['healthy', 'sick'])
```

>Note: specifying the labels will cause `healthy` to be the negative, `sick` the positive class.

## Metrics calculated from the confusion matrix
### Accuracy
$$
acc = \frac{TP + TN}{TP + TN + FP + FN}
$$
